# Exercise 4.2  
The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?

```python
import pandas as pd
import numpy.random as random
import thinkstats2
import thinkplot
```

```python
sample = random.random(1000)
random_sample_pmf = thinkstats2.Pmf(sample)
random_sample_cdf = thinkstats2.Cdf(sample)
```

# We expect a uniform PMF to show a horizontal line (meaning equal probability for all values), and the corresponding CDF to show a linear increase.

```python

thinkplot.Pmf(random_sample_pmf, label = 'PMF')
thinkplot.Show(xlabel='Probability distribution for random.random', 
               ylabel='PMF',
               ylim = [0., 0.002])


thinkplot.Cdf(random_sample_cdf, label = 'CDF')
thinkplot.Show(xlabel='Cumulative P distribution for random.random ', 
               ylabel='CDF')
```
